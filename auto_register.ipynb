## Import lib

```python
import requests
import pandas as pd
import lxml.html as LH
```

## Format Table View

```python
def find_courses(course, df):
    x = df.loc[df[4] == course, 'ID']
    return int(x)

def init_courses(a, drop=[0, 2, 3, 8, 10]):
    df_list = pd.read_html('<table>' + a.text + "</table>")[0]
    table = LH.fromstring(a.content)
    df_id = table.xpath('//tr/td/input/@data-rowindex')
    df_title = table.xpath('//tr/td[1]')
    df_id_all = []
    cnt = 0
    for df in df_title:
        if len(df) == 1:
            df_id_all.append(df_id[cnt])
            cnt += 1
        else:
            df_id_all.append("Full")
    df_list.insert(0, "ID", df_id_all, True)
    df_list[1] = df_list[1].str.replace(r'\s+', ' ', regex=True)
    df_list[4] = df_list[4].str.replace(r'\s+', ' ', regex=True)
    df_list = df_list[df_list.ID != 'Full']
    df_list = df_list.drop(drop, axis=1)
    return df_list

def init_courses_2(a, drop=[0, 2, 3, 8, 10]):
    df_list = pd.read_html('<table>' + a.text + "</table>")[0]
    table = LH.fromstring(a.content)
    df_id = table.xpath('//tr/td/input/@data-rowindex')
    df_title = table.xpath('//tr/td[1]')
    df_status = table.xpath('//tr/@class')
    df_id_all = []
    df_status_all = []
    cnt = 0
    for df in df_title:
        if len(df) == 1:
            df_id_all.append(df_id[cnt])
            cnt += 1
        else:
            df_id_all.append("Full")
    df_list.insert(0, "ID", df_id_all, True)

    for df in df_status:
        df_status_all.append(df)
    df_list.insert(1, "Status", df_status_all, True)
    
    df_list[1] = df_list[1].str.replace(r'\s+', ' ', regex=True)
    df_list[4] = df_list[4].str.replace(r'\s+', ' ', regex=True)
    df_list = df_list[df_list.ID != 'Full']
    df_list = df_list.drop(drop, axis=1)
    return df_list
```
